# ðŸŽ¨ BAY States Visualization - Quick Start Notebook

This notebook provides an interactive introduction to the BAY States Visualization System.

**States:**
- **B (Baseline)** - Stable, low-activation state
- **A (Activated)** - High-energy, fast-response state
- **Y (Yielding)** - Transitional, decaying state

## ðŸ“¦ Step 1: Import Libraries

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import the BAY States Analyzer
from bay_states_visualizations import BAYStatesAnalyzer

# Configure notebook
%matplotlib inline
plt.rcParams['figure.dpi'] = 100

print("âœ… All imports successful!")
```

## ðŸš€ Step 2: Create Analyzer Instance

```python
# Create analyzer with 1000 samples
analyzer = BAYStatesAnalyzer(n_samples=1000, random_state=42)

print("âœ… Analyzer created!")
print(f"ðŸ“Š Total samples: {len(analyzer.data)}")
print(f"ðŸŽ¯ States: {analyzer.data['State'].unique()}")
```

## ðŸ“Š Step 3: Explore the Data

```python
# View first few rows
print("ðŸ“‹ First 10 rows of data:\n")
display(analyzer.data.head(10))
```

```python
# Summary statistics
print("ðŸ“ˆ Summary Statistics:\n")
display(analyzer.data.groupby('State').describe())
```

```python
# State counts
print("ðŸ”¢ Samples per state:\n")
print(analyzer.data['State'].value_counts())

# Visualize distribution
plt.figure(figsize=(8, 6))
analyzer.data['State'].value_counts().plot(kind='bar', 
                                           color=['#2E86AB', '#A23B72', '#F18F01'])
plt.title('BAY States Distribution', fontweight='bold', fontsize=14)
plt.xlabel('State', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## ðŸ“Š Step 6: Custom Analysis

```python
# Create a custom scatter plot
plt.figure(figsize=(10, 8))

for state in ['B', 'A', 'Y']:
    data_state = analyzer.data[analyzer.data['State'] == state]
    plt.scatter(data_state['Activation'], 
               data_state['Energy'],
               label=f"{state} - {analyzer.bay_states[state]['name']}",
               c=analyzer.bay_states[state]['color'],
               s=100, alpha=0.6, edgecolors='black', linewidth=0.5)

plt.xlabel('Activation Level', fontweight='bold', fontsize=12)
plt.ylabel('Energy Consumption', fontweight='bold', fontsize=12)
plt.title('Custom Plot: Activation vs Energy', fontweight='bold', fontsize=14)
plt.legend(loc='best')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Statistical comparison
from scipy import stats

# Compare activation levels between states
B_activation = analyzer.data[analyzer.data['State'] == 'B']['Activation']
A_activation = analyzer.data[analyzer.data['State'] == 'A']['Activation']
Y_activation = analyzer.data[analyzer.data['State'] == 'Y']['Activation']

# ANOVA test
f_stat, p_value = stats.f_oneway(B_activation, A_activation, Y_activation)

print("\nðŸ“Š ANOVA Test Results:")
print(f"F-statistic: {f_stat:.2f}")
print(f"p-value: {p_value:.4e}")

if p_value < 0.001:
    print("âœ… Highly significant difference (p < 0.001)")
elif p_value < 0.05:
    print("âœ… Significant difference (p < 0.05)")
else:
    print("âŒ No significant difference")
```

## ðŸ“ˆ Step 7: Time Series Analysis

```python
# Plot time series for all states
plt.figure(figsize=(14, 6))

plt.plot(analyzer.time, analyzer.B_timeseries, 
        label='B - Baseline', color='#2E86AB', linewidth=2)
plt.plot(analyzer.time, analyzer.A_timeseries, 
        label='A - Activated', color='#A23B72', linewidth=2)
plt.plot(analyzer.time, analyzer.Y_timeseries, 
        label='Y - Yielding', color='#F18F01', linewidth=2)

plt.xlabel('Time (s)', fontweight='bold', fontsize=12)
plt.ylabel('Signal Amplitude', fontweight='bold', fontsize=12)
plt.title('BAY States: Temporal Dynamics', fontweight='bold', fontsize=14)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## ðŸŽ¯ Step 8: Feature Correlations

```python
# Correlation matrix for each state
features = ['Activation', 'Response_Time', 'Stability', 'Energy', 'Coherence']

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for idx, state in enumerate(['B', 'A', 'Y']):
    data_state = analyzer.data[analyzer.data['State'] == state][features]
    corr = data_state.corr()
    
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', 
               center=0, vmin=-1, vmax=1, square=True, 
               ax=axes[idx], cbar_kws={'shrink': 0.8})
    axes[idx].set_title(f'{state} - {analyzer.bay_states[state]["name"]}', 
                       fontweight='bold', fontsize=12)

plt.suptitle('Feature Correlations by State', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
